```python
import torch
import torch.nn as nn
```

```python
# Generator Architecture Explanation
#This generator upsamples a noise vector of size `latent_dim` (default 100) to a 224×224 RGB image.
#Uses ConvTranspose2d layers with BatchNorm and ReLU for stable training.
#Final layer uses Tanh activation to output pixel values scaled between -1 and 1.
#This directly generates 224x224 images, fulfilling the **resize to 224 x 224** requirement without external resizing.
```

```python
# Define the WGAN Generator architecture
class WGANGenerator(nn.Module):
    def __init__(self, latent_dim=100, img_channels=3, feature_maps=64):
        super(WGANGenerator, self).__init__()
        self.net = nn.Sequential(
            nn.ConvTranspose2d(latent_dim, feature_maps * 16, 7, 1, 0, bias=False),
            nn.BatchNorm2d(feature_maps * 16),
            nn.ReLU(True),
            nn.ConvTranspose2d(feature_maps * 16, feature_maps * 8, 4, 2, 1, bias=False),
            nn.BatchNorm2d(feature_maps * 8),
            nn.ReLU(True),
            nn.ConvTranspose2d(feature_maps * 8, feature_maps * 4, 4, 2, 1, bias=False),
            nn.BatchNorm2d(feature_maps * 4),
            nn.ReLU(True),
            nn.ConvTranspose2d(feature_maps * 4, feature_maps * 2, 4, 2, 1, bias=False),
            nn.BatchNorm2d(feature_maps * 2),
            nn.ReLU(True),
            nn.ConvTranspose2d(feature_maps * 2, feature_maps, 4, 2, 1, bias=False),
            nn.BatchNorm2d(feature_maps),
            nn.ReLU(True),
            nn.ConvTranspose2d(feature_maps, img_channels, 4, 2, 1, bias=False),
            nn.Tanh()
        )

    def forward(self, x):
        # Reshape input noise vector to (batch_size, latent_dim, 1, 1)
        x = x.view(x.size(0), x.size(1), 1, 1)
        return self.net(x)
```

```python
# Model Initialization
# Creates the generator model instance.
# Moves it to the GPU if available.
# Input Noise Creation
# Noise vectors sampled from a standard normal distribution.
# Shape: (batch_size, latent_dim).
```

```python
# Example usage:
if __name__ == "__main__":
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    latent_dim = 100
    batch_size = 8

    # Instantiate the generator
    gen = WGANGenerator(latent_dim=latent_dim).to(device)

    # Create a batch of random noise vectors
    noise = torch.randn(batch_size, latent_dim, device=device)

    # Generate images
    fake_images = gen(noise)

    print(f"Generated images shape: {fake_images.shape}")  # Expected: (8, 3, 224, 224)
```

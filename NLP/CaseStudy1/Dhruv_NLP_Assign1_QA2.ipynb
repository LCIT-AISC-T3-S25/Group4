```python
#Defining the regex pattern for words starting with 'w' or 'W'
pattern = r'\b[wW]\w*'
```

```python
#Using regex to find matches in each record
matches_w = df['Data'].str.contains(pattern, regex=True, na=False)
```

```python
#Counting the number of matching records
records_with_w_words = matches_w.sum()
```

```python
#Displaying the result
print(f"Number of records with a word starting with 'w' or 'W': {records_with_w_words}")
```

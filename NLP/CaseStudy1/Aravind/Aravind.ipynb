```python
# Clean the data (lowercase and strip) to avoid any conflicts
def clean_data(data):
    return data.str.lower().str.strip()

data_cleaned = clean_data(data['Data'])
```

```python
# 1. How many records have a date expressed without using alphabets?
date_pattern = r'\b\d{1,2}\/\d{1,2}\/\d{4}\b|\b\d{4}-\d{2}-\d{2}\b'  # Matching dates like dd/mm/yyyy or yyyy-mm-dd
dates_found = data_cleaned.str.contains(date_pattern).sum()
```

```python
print("Records with dates without alphabets", dates_found)
```

```python
import pandas as pd
import re
import nltk
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
# Download necessary NLTK data files
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
# Loading the dataset
file_path = './Data.csv'  
data = pd.read_csv(file_path)
```

```python
# Clean the data (lowercase and strip)
def clean_data(data):
    return data.str.lower().str.strip()

data_cleaned = clean_data(data['Data'])
```

```python
# Initialize the stemmer and lemmatizer
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
```

```python
import nltk
nltk.data.path.append('C:/Users/User/nltk_data')  # Adding the path where nltk_data is stored
```

```python
nltk.download('punkt_tab')  # Manually downloading the punkt_tab resource
```

```python
# Tokenizing the data for stemming and lemmatization
tokens = data_cleaned.apply(word_tokenize)
```

```python
# 2. Performing lemmatization
tokens_lemmatized = tokens.apply(lambda x: [lemmatizer.lemmatize(word) for word in x])

# Counting the unique words before and after lemmatization
unique_words_before_lemmatization = set(tokens.explode())
unique_words_after_lemmatization = set(tokens_lemmatized.explode())
print("Unique words before lemmatization:", len(unique_words_before_lemmatization))
print("Unique words after lemmatization:", len(unique_words_after_lemmatization))
```

# 5. How many records contain a decimal number?

```python
import pandas as pd
import re
```

```python
# Load the CSV file
df = pd.read_csv("Data.csv", header=None, names=["tweet"])
```

```python
# Step 1: Remove duplicate rows
df_unique = df.drop_duplicates()
```

```python
# Step 2: Convert all data to string format for regex application
df_str = df_unique.astype(str)
```

```python
# Step 3: Define regex pattern to match decimal numbers (e.g., 3.14)
decimal_pattern = r'\b\d+\.\d+\b'
```

```python
# Step 4: Identify rows that contain at least one decimal using regex
decimal_mask = df_str.apply(lambda row: row.str.contains(decimal_pattern, regex=True)).any(axis=1)
```

```python
# Step 5: Apply pattern to each tweet and count how many contain decimal numbers
decimal_count = df['tweet'].apply(lambda x: bool(re.search(decimal_pattern, str(x)))).sum()
```

```python
print("Number of records containing a decimal number:", decimal_count)
```

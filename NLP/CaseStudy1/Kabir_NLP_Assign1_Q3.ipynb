```python
# Define regex pattern for matching words starting with an alphabet (a-z or A-Z), not starting with 'http' or 'www'
pattern = r'\b(?!http)(?!www)[a-zA-Z]\w*'
```

```python
# Function to check if any matching word exists in a record
def has_valid_word(text):
    if pd.isnull(text):
        return False
    return any(re.match(pattern, word) for word in str(text).split())
```

```python
# Apply function to all rows in relevant column(s)
column_to_check = 'Data'  # Replace with your actual column name
df['has_valid_word'] = df[column_to_check].apply(has_valid_word)
```

```python
# Count how many records contain a valid word
count = df['has_valid_word'].sum()
print(f"Number of records with a word that starts with an alphabet and is not a URL: {count}")
```

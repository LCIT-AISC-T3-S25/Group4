```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from collections import Counter
```

```python
# Download required resources
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
# Load data
df = pd.read_csv("Data.csv", header=None, names=["tweet"])
```

```python
# Step 1: Combine all tweets into one text
text = " ".join(df["tweet"].dropna().astype(str)).lower()
```

```python
# Step 2: Tokenize
tokens = word_tokenize(text)
```

```python
# Step 3: Remove stopwords
stop_words = set(stopwords.words("english"))
tokens_no_stop = [word for word in tokens if word.isalpha() and word not in stop_words]
```

```python
# Step 4a: Stemming (normalization after stop word removal)
stemmer = PorterStemmer()
stemmed_tokens = [stemmer.stem(word) for word in tokens_no_stop]
stemmed_freq = Counter(stemmed_tokens)
```

```python
# Step 4b: Lemmatization (normalization after stop word removal)
lemmatizer = WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(word) for word in tokens_no_stop]
lemmatized_freq = Counter(lemmatized_tokens)
```

```python
# Step 5: Print results
# Convert top 10 stemmed words to a DataFrame
stem_df = pd.DataFrame(stemmed_freq.most_common(10), columns=["Word (Stemmed)", "Frequency"])
print("\n🔹 Top 10 Word Frequencies After Stemming (post stopword removal):")
print(stem_df.to_string(index=False))

# Convert top 10 lemmatized words to a DataFrame
lemma_df = pd.DataFrame(lemmatized_freq.most_common(10), columns=["Word (Lemmatized)", "Frequency"])
print("\n🔹 Top 10 Word Frequencies After Lemmatization (post stopword removal):")
print(lemma_df.to_string(index=False))
```

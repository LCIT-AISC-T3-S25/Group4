```python
import pandas as pd
```

```python
# Loading the dataset
data = pd.read_csv('C:/Users/nidhi/OneDrive/Documents/Data.csv') 
```

```python
# Clean the data: lowercase and strip
def clean_data(column):
    return column.str.lower().str.strip()
data_cleaned = clean_data(data['Data'])
```

```python
newline_pattern = r'\n'
newline_count = data_cleaned.astype(str).str.contains(newline_pattern).sum()
print(f"Records with a newline character: {newline_count}")
```

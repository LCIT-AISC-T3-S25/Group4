```python
import pandas as pd
import re
from google.colab import files

# Prompt user to upload a file
uploaded = files.upload()



```

```python
# Assuming the uploaded file is a CSV and there's only one file uploaded
# Get the filename from the uploaded dictionary
filename = list(uploaded.keys())[0]

# Read the uploaded file into a pandas DataFrame
df = pd.read_csv(filename)

# Display the first few rows to understand the structure
print("Columns:", df.columns)
df.head()
```

```python
tweet_column = 'Data'
```

```python



#  specified tweet column
tweet_column = 'Data'  # This should match the actual column name with tweets

#  Define hashtag counting function
def count_hashtags(text):
    if pd.isnull(text):
        return 0
    return len(re.findall(r'#\w+', text))

#  Apply function and compute total
df['hashtag_count'] = df[tweet_column].apply(count_hashtags)
total_hashtags = df['hashtag_count'].sum()

#  Print result
print("✅ Total number of hashtags across all tweets:", total_hashtags)
```

```python
text_column = None
for col in df.columns:
    if df[col].dtype == 'object' and df[col].astype(str).str.contains('#').any():
        text_column = col
        break

if text_column is None:
    raise ValueError("❌ No column with hashtags found. Please check your Excel data.")

print(f"📄 Detected text column for hashtag analysis: {text_column}")

# 📌 Step 4: Count Hashtags
def count_hashtags(text):
    if pd.isnull(text):
        return 0
    return len(re.findall(r'#\w+', str(text)))

# Apply and sum
df['hashtag_count'] = df[text_column].apply(count_hashtags)
total_hashtags = df['hashtag_count'].sum()

print(f"✅ Total number of hashtags in '{text_column}':", total_hashtags)

# 📌 Step 5 (Optional): Show Top Hashtags by Frequency
from collections import Counter

def extract_hashtags(text):
    return re.findall(r'#\w+', str(text).lower())

all_hashtags = [ht for text in df[text_column] for ht in extract_hashtags(text)]
hashtag_freq = Counter(all_hashtags).most_common(10)

print("\n🔝 Top 10 Hashtags:")
for tag, count in hashtag_freq:
    print(f"{tag}: {count}")
```

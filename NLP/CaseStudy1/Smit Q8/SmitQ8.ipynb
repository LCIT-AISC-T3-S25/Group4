```python
import pandas as pd
import re
from collections import Counter

# ✅ Step 1: Detect the Text Column Containing Hashtags
text_column = None
for col in df.columns:
    if df[col].dtype == 'object' and df[col].astype(str).str.contains(r'#\w+', na=False).any():
        text_column = col
        break

if text_column is None:
    raise ValueError("❌ No column with hashtags found. Please check your Excel data.")

print(f"📄 Detected text column for hashtag analysis: {text_column}")

# ✅ Step 2: Count Hashtags per Row
def count_hashtags(text):
    if pd.isnull(text):
        return 0
    return len(re.findall(r'#\w+', str(text)))

df['hashtag_count'] = df[text_column].apply(count_hashtags)
total_hashtags = df['hashtag_count'].sum()
print(f"✅ Total number of hashtags in '{text_column}': {total_hashtags}")

# ✅ Step 3: Extract All Hashtags and Count Unique Ones
def extract_hashtags(text):
    return re.findall(r'#\w+', str(text).lower())

# Flatten all hashtags into a single list
all_hashtags = [hashtag for text in df[text_column] if pd.notnull(text) for hashtag in extract_hashtags(text)]

# Total unique hashtags
unique_hashtags = set(all_hashtags)
print(f"🔢 Total number of unique hashtags: {len(unique_hashtags)}")

# ✅ Step 4: Show Top 10 Hashtags by Frequency
hashtag_freq = Counter(all_hashtags).most_common(10)
print("\n🔝 Top 10 Hashtags:")
for tag, count in hashtag_freq:
    print(f"{tag}: {count}")
```

```python
# Function to count URLs in text
def count_urls(text):
    url_pattern = r"http\S+|www\S+"
    return len(re.findall(url_pattern, str(text)))

# Apply the function to count URLs in each tweet
df['url_count'] = df[tweet_column].apply(count_urls)

# Sum total URLs
total_urls = df['url_count'].sum()

# Print result
print(f"🔗 Total URLs found: {total_urls}")
```

```python
import re

# Function to extract URLs from text
def extract_urls(text):
    url_pattern = r"http\S+|www\S+"
    return re.findall(url_pattern, str(text))

# Apply the function to extract URLs from each row
df['extracted_urls'] = df[tweet_column].apply(extract_urls)

# Flatten all URLs into a single list
all_urls = [url for sublist in df['extracted_urls'] for url in sublist]

# Count total
total_urls = len(all_urls)

# Print total and URLs
print(f"🔗 Total URLs found: {total_urls}")
print("🔗 List of URLs:")
for url in all_urls:
    print(url)
```

```python
import re

# Improved function to extract URLs
def extract_valid_urls(text):
    url_pattern = r"https?://[^\s]+|www\.[a-zA-Z0-9\-]+\.[a-zA-Z]{2,}"
    return re.findall(url_pattern, str(text))

# Apply to DataFrame
df['extracted_urls'] = df[tweet_column].apply(extract_valid_urls)

# Flatten all URLs into a single list
all_urls = [url for sublist in df['extracted_urls'] for url in sublist]

# Count and print
total_urls = len(all_urls)
print(f"🔗 Total Valid URLs found: {total_urls}")
print("🔗 List of Valid URLs:")
for url in all_urls:
    print(url)
```

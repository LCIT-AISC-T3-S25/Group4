```python
import pandas as pd
import nltk
import re
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer


# Download tokenizer
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
df = pd.read_csv("Data-2.csv")
```

```python
# Combine all text into one string
all_text = " ".join(df['Data'].dropna().astype(str))

# Tokenize and clean (keep only alphabetic tokens, lowercase)
tokens = word_tokenize(all_text)
tokens = [token.lower() for token in tokens if token.isalpha()]

# Count unique tokens before lemmatization
unique_before = set(tokens)

# Initialize lemmatizer
lemmatizer = WordNetLemmatizer()

# Lemmatize tokens
lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]

# Count unique tokens after lemmatization
unique_after = set(lemmatized_tokens)

# Print results
print(f"Unique words before lemmatization: {len(unique_before)}")
print(f"Unique words after lemmatization:  {len(unique_after)}")
```

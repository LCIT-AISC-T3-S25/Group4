```python
import re

# Load your dataset
df = pd.read_csv("Data-2.csv")

# Define a function to replace non-alphanumeric characters with newlines
def replace_non_alnum_with_newline(text):
    return re.sub(r'[^a-zA-Z0-9]', '\n', str(text))

# Apply the function to the 'Data' column
df['Processed_Data'] = df['Data'].apply(replace_non_alnum_with_newline)

# Example: print first 3 processed rows
print(df['Processed_Data'].head(3))
```

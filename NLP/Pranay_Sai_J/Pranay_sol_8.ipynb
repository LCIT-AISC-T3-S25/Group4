```python
import pandas as pd
import re
from collections import Counter

# Load data
df = pd.read_csv("Data.csv", low_memory=False)

# Assume the only column contains the tweets
text_column = df.columns[0]

# Function to extract hashtags from a string
def extract_hashtags(text):
    return re.findall(r'#\w+', str(text))

# Extract and flatten all hashtags
all_hashtags = df[text_column].apply(extract_hashtags).sum()

# Count hashtag frequencies
hashtag_counts = Counter(all_hashtags)

print(f"Total number of hashtags across all tweets: {total_hashtags}")

# Get all hashtags
all_hashtags = hashtag_counts.most_common(total_hashtags)

# Display results
print("All Hashtags:")
for tag, count in all_hashtags:
    print(f"{tag}: {count}")
```

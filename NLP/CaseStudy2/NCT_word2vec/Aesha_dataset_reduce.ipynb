```python
import pandas as pd

# Load original datasets
train_df = pd.read_csv("../cleaned_train_balanced.csv")
test_df = pd.read_csv("../cleaned_test.csv")

# Reduce train to 10,000 samples (stratified by 'sentiment')
train_df_reduced = train_df.groupby("sentiment", group_keys=False).apply(
    lambda x: x.sample(frac=10000 / len(train_df), random_state=42)
).reset_index(drop=True)

# Reduce test to 10,000 samples (stratified by 'sentiment')
test_df_reduced = test_df.groupby("sentiment", group_keys=False).apply(
    lambda x: x.sample(frac=10000 / len(test_df), random_state=42)
).reset_index(drop=True)

# Print size and class balance
print(f"Train reduced shape: {train_df_reduced.shape}")
print(f"Test reduced shape: {test_df_reduced.shape}")
print("\nTrain class distribution:\n", train_df_reduced['sentiment'].value_counts())
print("\nTest class distribution:\n", test_df_reduced['sentiment'].value_counts())

# Save reduced files
train_df_reduced.to_csv("cleaned_train_balanced_reduced.csv", index=False)
test_df_reduced.to_csv("cleaned_test_reduced.csv", index=False)
```
